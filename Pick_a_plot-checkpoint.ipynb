## Insert images at a given position in a plot

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# generate random data
from random import random
a =[(2+random()) ** 5 for _ in range(10)]

# load some image
import urllib
URL='http://img3.wikia.nocookie.net/__cb20120301204636/tinyzoo/images/6/6a/Naked_mole_rat_single.png'
imgfile = urllib.urlopen(URL)

# plot
img = plt.imread(imgfile)
fig, ax = plt.subplots()
ax.bar(range(10), a, color='seagreen')
xlim = ax.get_xlim()
xdiff = (xlim[1] - xlim[0])
ylim = ax.get_ylim()
yscale = (ylim[1] - ylim[0]-ylim[1]/11)/xdiff
for i in range(10):
    ax.imshow(img, extent=[i, i+(xdiff/11), a[i], a[i]+yscale],
              aspect='auto', cmap='gray')
ax.set_ylim((ylim[0], ylim[1]+yscale))
ax.set_xlim(xlim)
plt.show()
```
